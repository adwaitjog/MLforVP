```python
comparison = pd.DataFrame(columns=['Error'])
```

```python
comparison.loc['Regular OLS'] = regular_ols['Error'].drop('compute_int_15-loads.csv').mean()
comparison.loc['Split OLS'] = split_ols['Error'].drop('compute_int_15-loads-split.csv').mean()
comparison.loc['OLS 16'] = ols_16['Error'].drop('compute_int_15-loads.csv').mean()
comparison.loc['Regular DNN'] = regular_dnn['Error'].drop('compute_int_15-loads.csv').mean()
comparison.loc['Split DNN'] = split_dnn['Error'].drop('compute_int_15-loads-split.csv').mean()
comparison.loc['CNN'] = cnn['Error'].drop('compute_int_15-loads.csv').mean()
```
