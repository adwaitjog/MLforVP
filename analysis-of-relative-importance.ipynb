```python
import pandas as pd

regular_ols = pd.read_csv('ols-pc-eff-lvp-RESULTS.csv', index_col=0)
split_ols   = pd.read_csv('ols-split-RESULTS.csv', index_col=0)
ols_16      = pd.read_csv('ols16-lvpRESULTS.csv', index_col=0)
```

```python
# Check the predictive power of the pc vs effective address
regular_ols[['pc', 'effective_address']].plot.bar(logy=True, title='Importance of PC vs Effective address')
```

```python
# Check the predictive power of the PC and Effective address in the split ols model
low_network = split_ols[["low_pc_high", "low_effective_address_low", "low_effective_address_high"]]
mapping = {"low_pc_high" : "low_pc", "low_effective_address_low" : "low_effective_address", "low_effective_address_high" : "high_effective_address"}
low_network = low_network.rename(columns=mapping)
low_network.plot.bar(logy=True, title='Importance of PC vs Effective address')
```
